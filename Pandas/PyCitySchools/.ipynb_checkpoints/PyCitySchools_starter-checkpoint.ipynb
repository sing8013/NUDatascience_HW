# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
dist_sum_df=pd.DataFrame({'Total Schools' : len(school_data['school_name']),
                          'Total Students': len(student_data['Student ID']),
                          'Total Budget': '${0:,}'.format(school_data['budget'].sum()),
                          'Average Math Score': '{0:.2f}'.format(school_data_complete['math_score'].mean()),
                          'Average Reading Score': '{0:.2f}'.format(school_data_complete['reading_score'].mean()),
                          'Passing Math	%': '{0:.2f}'.format(len(school_data_complete.loc[school_data_complete['math_score']>=70])/len(student_data['Student ID'])*100),
                          'Passing Reading	%': '{0:.2f}'.format(len(school_data_complete.loc[school_data_complete['reading_score']>=70])/len(student_data['Student ID'])*100),     
                         },index=[0])
#Overall Passing Rate
dist_sum_df['Overall Passing Rate']='{0:.2f}'.format((float(dist_sum_df['Passing Math	%'])+ float(dist_sum_df['Passing Reading	%']))/2)
dist_sum_df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#group complete DataFrame by School
school_data_complete.set_index('school_name')
group_df=school_data_complete.groupby(['school_name'])
#group_df.set_index('school_name')

school_df=pd.DataFrame({'School Type': group_df['type'].nth(0),
                        'Total Students': group_df['size'].count(),
                        'Total School Budget': group_df['budget'].nth(0),
                        'Per Student Budget': group_df['budget'].nth(0)/group_df['size'].nth(1),
                        'Average Math Score': group_df['math_score'].agg('mean').map('{0:.2f}'.format),
                        'Average Reading Score': group_df['reading_score'].agg('mean').map('{0:.2f}'.format),
                        'Passing Math	%': (100*school_data_complete.loc[school_data_complete['math_score']>=70].groupby(['school_name']).count()['math_score']/group_df['size'].count()).map('{0:.2f}'.format),
                        'Passing Reading	%': (100*school_data_complete.loc[school_data_complete['reading_score']>=70].groupby(['school_name']).count()['math_score']/group_df['size'].count()).map('{0:.2f}'.format),     
                        })
#Overall Passing Rate
school_df['Overall Passing Rate']=((school_df['Passing Math	%'].astype(float)+ school_df['Passing Reading	%'].astype(float))/2).map('{0:.2f}'.format)
#result_df
# Sort and display the top five schools in overall passing rate
result_df=school_df.sort_values(by='Overall Passing Rate',ascending=False)
result_df.head(5)
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
# Sort and display the five wors-performing schools in overall passing rate
result_df=school_df.sort_values(by='Overall Passing Rate',ascending=True)
result_df.head(5)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Merge school_df to the school_data_complete df on school_name or school_ID
merge_school_df=pd.merge(school_data_complete, school_df, how="left", on=["school_name", "school_name"])
#use cut function to split data into bins
merge_school_df['Spending Ranges (Per Student)']=pd.cut(merge_school_df['Per Student Budget'],spending_bins,labels=group_names)

#group complete DataFrame by Spending Ranges
merge_school_df.set_index('Spending Ranges (Per Student)')
group_df=merge_school_df.groupby(['Spending Ranges (Per Student)'])

result_df=pd.DataFrame({'Average Math Score': group_df['math_score'].agg('mean').map('{0:.2f}'.format),
                        'Average Reading Score': group_df['reading_score'].agg('mean').map('{0:.2f}'.format),
                        'Passing Math	%': (100*merge_school_df.loc[merge_school_df['math_score']>=70].groupby(['Spending Ranges (Per Student)']).count()['math_score']/group_df['size'].count()).map('{0:.2f}'.format),
                        'Passing Reading	%': (100*merge_school_df.loc[merge_school_df['reading_score']>=70].groupby(['Spending Ranges (Per Student)']).count()['reading_score']/group_df['size'].count()).map('{0:.2f}'.format),     
                        })
#Overall Passing Rate
result_df['Overall Passing Rate']=((result_df['Passing Math	%'].astype(float)+ result_df['Passing Reading	%'].astype(float))/2).map('{0:.2f}'.format)
result_df.head()
#merge_school_df.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
#Merge school_df to the school_data_complete df on school_name or school_ID
merge_school_df=pd.merge(school_data_complete, school_df, how="left", on=["school_name", "school_name"])
#use cut function to split data into bins
merge_school_df['School Size']=pd.cut(merge_school_df['size'],size_bins,labels=group_names)

#group complete DataFrame by Spending Ranges
merge_school_df.set_index('School Size')
group_df=merge_school_df.groupby(['School Size'])

result_df=pd.DataFrame({'Average Math Score': group_df['math_score'].agg('mean').map('{0:.2f}'.format),
                        'Average Reading Score': group_df['reading_score'].agg('mean').map('{0:.2f}'.format),
                        'Passing Math	%': (100*merge_school_df.loc[merge_school_df['math_score']>=70].groupby(['School Size']).count()['math_score']/group_df['size'].count()).map('{0:.2f}'.format),
                        'Passing Reading	%': (100*merge_school_df.loc[merge_school_df['reading_score']>=70].groupby(['School Size']).count()['reading_score']/group_df['size'].count()).map('{0:.2f}'.format),     
                        })
#Overall Passing Rate
result_df['Overall Passing Rate']=((result_df['Passing Math	%'].astype(float)+ result_df['Passing Reading	%'].astype(float))/2).map('{0:.2f}'.format)
result_df.head()
```

* Perform the same operations as above, based on school type.

```python
#group complete DataFrame by Type
merge_school_df.set_index('type')
group_df=merge_school_df.groupby(['type'])

result_df=pd.DataFrame({'Average Math Score': group_df['math_score'].agg('mean').map('{0:.2f}'.format),
                        'Average Reading Score': group_df['reading_score'].agg('mean').map('{0:.2f}'.format),
                        'Passing Math	%': (100*merge_school_df.loc[merge_school_df['math_score']>=70].groupby(['School Type']).count()['math_score']/group_df['size'].count()).map('{0:.2f}'.format),
                        'Passing Reading	%': (100*merge_school_df.loc[merge_school_df['reading_score']>=70].groupby(['School Type']).count()['reading_score']/group_df['size'].count()).map('{0:.2f}'.format),     
                        })
#Overall Passing Rate
result_df['Overall Passing Rate']=((result_df['Passing Math	%'].astype(float)+ result_df['Passing Reading	%'].astype(float))/2).map('{0:.2f}'.format)
result_df.head()
```
