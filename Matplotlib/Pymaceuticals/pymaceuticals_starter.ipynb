```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df=pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
merge_df=pd.merge(clinical_trial_df,mouse_drug_df,how='left')

# Display the data table for preview
merge_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df=merge_df.groupby(['Drug','Timepoint'])
mean_tumor_volume=grouped_df['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
mean_tumor_volume=pd.DataFrame(mean_tumor_volume)
# Preview DataFrame
mean_tumor_volume=mean_tumor_volume.reset_index()
mean_tumor_volume.head(11)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_tumor_volume=grouped_df['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
sem_tumor_volume=pd.DataFrame(sem_tumor_volume)
# Preview DataFrame
sem_tumor_volume=sem_tumor_volume.reset_index()
sem_tumor_volume.head()
```

```python
#Mean Volume
mean_vol_df = pd.pivot(index=mean_tumor_volume.Timepoint, columns=mean_tumor_volume.Drug, values=mean_tumor_volume['Tumor Volume (mm3)'])
#mean_df=mean_df.reset_index()
#Standard Error
sem_df=pd.pivot(index=sem_tumor_volume.Timepoint, columns=sem_tumor_volume.Drug, values=sem_tumor_volume['Tumor Volume (mm3)'])
#sem_df=sem_df.reset_index()
# Preview that Reformatting worked
mean_vol_df.head()
#sem_df.head()
```

```python
x_axis=mean_vol_df[mean_vol_df.columns[0]]
# Generate the Plot (with Error Bars)
col='Capomulin'
#for col in mean_df.columns:
plt.errorbar(x_axis,mean_vol_df[col],yerr=sem_df[col],fmt='',label=col,marker='o',linestyle='--')

col='Infubinol'
plt.errorbar(x_axis,mean_vol_df[col],yerr=sem_df[col],fmt='',label=col,marker='*',linestyle='--')

col='Ketapril'
plt.errorbar(x_axis,mean_vol_df[col],yerr=sem_df[col],fmt='',label=col,marker='s',linestyle='--')

col='Placebo'
plt.errorbar(x_axis,mean_vol_df[col],yerr=sem_df[col],fmt='',label=col,marker='d',linestyle='--')

plt.grid()
plt.title('Tumor Response to Treatment')
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.legend(loc='best')
# Save the Figure
plt.savefig('../Images/tumor_volume.png')
```

```python
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_mice=grouped_df['Mouse ID'].count()
# Convert to DataFrame
count_mice=pd.DataFrame(count_mice)
count_mice=count_mice.rename(columns={'Mouse ID':'Mouse Count'})
# Preview DataFrame
count_mice=count_mice.reset_index()
count_mice.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_df = pd.pivot(index=count_mice.Timepoint, columns=count_mice.Drug, values=count_mice['Mouse Count'])
# Preview that Reformatting worked
#count_df=count_df.reset_index()
count_df.head()
```

```python
#Initial count
first_row = count_df.iloc[0][:]
#Generate data frame of survival rates
surv_df=[row/first_row*100 for j, row in count_df.iterrows()]

surv_df=pd.DataFrame(surv_df)
surv_df['Timepoint']=count_df.index
surv_df=surv_df.set_index('Timepoint')
#count_df.index[0:]
surv_df
```

```python
# Generate the Plot (Accounting for percentages)
x_axis=surv_df.index.values
# Generate the Plot (with line plot)
col='Capomulin'
plt.plot(x_axis,surv_df[col],label=col,marker='o',linestyle='--')

col='Infubinol'
plt.errorbar(x_axis,surv_df[col],label=col,marker='*',linestyle='--')

col='Ketapril'
plt.errorbar(x_axis,surv_df[col],label=col,marker='s',linestyle='--')

col='Placebo'
plt.errorbar(x_axis,surv_df[col],label=col,marker='d',linestyle='--')

plt.grid()
plt.title('Survival During Treatment')
plt.xlabel('Time(Days)')
plt.ylabel('Survival Rate (%)')
plt.legend(loc='best')
# Save the Figure
plt.savefig('../Images/survival_rate.png')
```

## Summary Bar Graph

```python
# Calculate the percent changes (in Tumor Volume) for each drug
#Initial count
start_row = mean_vol_df.iloc[0][:]
end_row=mean_vol_df.iloc[len(mean_vol_df)-1][:]
chg_l=(end_row/start_row-1)*100
# Display the data to confirm
chg_l
```

```python
# Store all Relevant Percent Changes into a Tuple
chg_t=tuple(zip(chg_l.index,chg_l))

# Splice the data between passing and failing drugs
passed_drug=pd.DataFrame([x for x in chg_t if x[1]>=0])
failed_drug=pd.DataFrame([x for x in chg_t if x[1]<0])

splice_l={'pass':[x for x in chg_t if x[1]>=0],'fail':[x for x in chg_t if x[1]<0]}

# Orient widths. Add labels, tick marks, etc. 

width=0.75

#  create the figure
fig, ax = plt.subplots()
ax.set_ylabel('Percentage Change')
ax.set_title('Drug Impact during treatment')

#ax.legend()

plt.axhline(0)
plt.grid()
plt.xticks()

# Use functions to label the percentages of changes
def add_text(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d%%' % int(height),
                ha='center', va='bottom')

def plot_drug(drug_name):   
    chg_val = [x[1] for x in chg_t if x[0]==drug_name]
    if chg_val[0]>=0:
        clr='g'
    else:
        clr='r'
    plot=ax.bar(drug_name,height=chg_val,color=clr,label=str(chg_val[0]),width=width)
    add_text(plot)

# Call functions to implement the function calls
plot_drug('Capomulin')
plot_drug('Infubinol')
plot_drug('Ketapril')
plot_drug('Placebo')

# Save the Figure
fig.savefig('../Images/bar_chart.png')

# Show the Figure
fig.show()
```
